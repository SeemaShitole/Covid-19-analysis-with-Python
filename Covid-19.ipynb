```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
Web Scrapping cleaning of Covid-19 data of World:

source: https://www.worldometers.info/coronavirus
```

```python
url="https://www.worldometers.info/coronavirus/"
```

```python
response=requests.get(url)
```

```python
print(response)
```

```python
soup=BeautifulSoup(response.content,"html.parser")
```

```python
soup.title
```

```python
divs=soup.find("div",class_="panel_flip")
for i in divs.find_all("div" , class_="panel_front"):
    print(i.text)
```

```python
coronatable=soup.find_all("table")
```

```python
len(coronatable)
```

```python
co=coronatable[0]
```

```python
country=[]
total_cases=[]
new_cases=[]
total_deaths=[]
new_deaths=[]
total_recovered=[]
active_cases=[]
serious_critical=[]
tot_cases_1M_pop=[]
deaths_1M_pop=[]
total_tests=[]
tests_1M_pop=[]
population=[]


rows=co.find_all("tr")[9:-8]
for row in rows:
    cell=row.find_all("td")
    country.append(cell[1].text.strip())
    total_cases.append(cell[2].text.strip())
    new_cases.append(cell[3].text.strip())
    total_deaths.append(cell[4].text.strip())
    new_deaths.append(cell[5].text.strip())
    total_recovered.append(cell[6].text.strip())
    active_cases.append(cell[7].text.strip())
    serious_critical.append(cell[8].text.strip())
    tot_cases_1M_pop.append(cell[9].text.strip())
    deaths_1M_pop.append(cell[10].text.strip())
    total_tests.append(cell[11].text.strip())
    tests_1M_pop.append(cell[12].text.strip())
    population.append(cell[13].text.strip())
    
    
print(country)
print(total_cases)
print(new_cases)
print(total_deaths)
print(new_deaths)
print(total_recovered)
print(active_cases)
print(serious_critical)
print(tot_cases_1M_pop)
print(deaths_1M_pop)
print(total_tests)
print(tests_1M_pop)
print(population)
    
```

```python
df=pd.DataFrame(list(zip(country, total_cases, new_cases,total_deaths,new_deaths,total_recovered,active_cases,serious_critical,tot_cases_1M_pop,deaths_1M_pop,total_tests,tests_1M_pop,population)), columns=["Country", "Total_Cases", "New_Cases","Total_Deaths","New_Deaths","Total_Recovered","Active_Cases","Serious_Critical","Total_Cases_1M_Pop","Deaths_1M_Pop","Total_Tests","Tests_1M_Pop","Population"])
df
```

```python
df.replace(r'^\s*$' ,np.nan, regex=True, inplace=True)
df.replace(np.nan,0, inplace=True)
df
```

```python


#Removing special characters from the col 

total_cases1=[]
for i in df["Total_Cases"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    total_cases1.append(s2)
    
# print(total_cases1)


new_cases1=[]
for i in df["New_Cases"]:
    s=str(i)
    s1=s.replace("+","")
    s2=s1.replace(",","")
    s3=int(s2)
    new_cases1.append(s3)
    
    
# print(new_cases1)


total_deaths1=[]
for i in df["Total_Deaths"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    total_deaths1.append(s2)
    
# print(total_deaths1)


new_deaths1=[]
for i in df["New_Deaths"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    new_deaths1.append(s2)
    
# print(new_deaths1)


total_recovered1=[]
for i in df["Total_Recovered"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    total_recovered1.append(s2)
    
# print(total_recovered1)


# Replace the N/A values with zero in active_cases1 col
df.replace('N/A',np.nan,inplace=True)
df.replace(np.nan,0,inplace=True)


active_cases1=[]
for i in df["Active_Cases"]:
    s=str(i)
    s1=s.replace("+","")
    s2=s1.replace(",","")
    s3=int(s2)
    active_cases1.append(s3)
    
    
# print(active_cases1)


serious_critical1=[]
for i in df["Serious_Critical"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    serious_critical1.append(s2)
    
# print(serious_critical1)
    
total_cases_1m_pop1=[]
for i in df["Total_Cases_1M_Pop"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    total_cases_1m_pop1.append(s2)
    
# print(total_cases_1m_pop1)
    
deaths_1m_pop1=[]
for i in df["Deaths_1M_Pop"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    deaths_1m_pop1.append(s2)
    
# print(deaths_1m_pop1)
    
# Total_Tests has no commas only 3 didgits in col
    
    
tests_1m_pop1=[]
for i in df["Tests_1M_Pop"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    tests_1m_pop1.append(s2)
    
# print(tests_1m_pop1)
    
    
population1=[]
for i in df["Population"]:
    s=str(i)
    s1=s.replace(",","")
    s2=int(s1)
    population1.append(s2)
    
# print(population1)
```

```python
df=pd.DataFrame(list(zip(country, total_cases1, new_cases1,total_deaths1,new_deaths1,total_recovered1,active_cases1,serious_critical1,total_cases_1m_pop1,deaths_1m_pop1,total_tests,tests_1m_pop1,population1)), columns=["Country", "Total_Cases", "New_Cases","Total_Deaths","New_Deaths","Total_Recovered","Active_Cases","Serious_Critical","Total_Cases_1M_Pop","Deaths_1M_Pop","Total_Tests","Tests_1M_Pop","Population"])
df
```

Web Scrapping cleaning of Covid-19 data for India:

source: https://www.mohfw.gov.in

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

```

```python
url=" https://www.mohfw.gov.in/"
```

```python
response=requests.get(url)
```

```python
print(response)
```

```python
soup=BeautifulSoup(response.content, "html.parser")
```

```python
soup.title
```

```python
coronatable_india=soup.find_all("table")
```

```python
print(coronatable_india)
```

```python
co1=coronatable_india[0]
```

```python
states=[]
total_cases=[]
cured_discharged_migrated=[]
deaths=[]

rows=co1.find_all('tr')[1:-6]
for row in rows:
    cell=row.find_all('td')
    states.append(cell[1].text.strip())
    total_cases.append(cell[2].text.strip())
    cured_discharged_migrated.append(cell[3].text.strip())
    deaths.append(cell[4].text.strip())
    
print(states)
print(total_cases)
print(cured_discharged_migrated)
print(deaths)
```

```python
df=pd.DataFrame(list(zip(states, total_cases, cured_discharged_migrated,deaths)), columns=["States", "Total_Cases", "Cured_Discharged_Migrated","Deaths"])
df
```
